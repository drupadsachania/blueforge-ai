```python
# Telemetry writer for local factory monitoring
import json, time, os
TELEM_DIR = f"{BASE}/telemetry"
os.makedirs(TELEM_DIR, exist_ok=True)
TELEM_FILE = f"{TELEM_DIR}/{RUN_ID}_telemetry.json"

def write_telemetry(status, phase="training", step=0, loss=None, gpu_minutes=0.0, artifact=""):
    payload = {
        "run_id": RUN_ID,
        "status": status,
        "phase": phase,
        "step": int(step),
        "loss": None if loss is None else float(loss),
        "gpu_minutes": float(gpu_minutes),
        "artifact": artifact,
        "ts": time.strftime("%Y-%m-%dT%H:%M:%SZ", time.gmtime())
    }
    with open(TELEM_FILE, "w", encoding="utf-8") as f:
        json.dump(payload, f)
    print("Telemetry updated:", payload)

# Example milestones:
# write_telemetry(status="training", step=100, loss=1.23, gpu_minutes=12.5)
# write_telemetry(status="training_complete", phase="export", gpu_minutes=42.0, artifact="soc-architect-qwen35-4b-q4_k_m.gguf")
```
